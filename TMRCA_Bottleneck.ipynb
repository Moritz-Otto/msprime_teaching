# TMRCA and Bottlenecks

Import dependencies

```python
import msprime
from IPython.display import SVG, display
```

```python
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

Parameters

```python
n = 30
N = 1e4

T1 = 1000
N1 = 500
T2 = 2000
N2 = 2000

demographic_events = [
    msprime.PopulationParametersChange(time=T1,initial_size=N1),
    msprime.PopulationParametersChange(time=T2,initial_size=N2),
]
ts = msprime.simulate(n,Ne=N, demographic_events=demographic_events)
tree = ts.first()
tree.draw_svg(y_axis=True,size=(600,800))
```

## 10,000 simulations

```python
n = 30
N = 1e4

T1 = 4000
N1 = 1500
T2 = 6500
N2 = 5000
demographic_events = [
    msprime.PopulationParametersChange(time=T1,initial_size=N1),
    msprime.PopulationParametersChange(time=T2,initial_size=N2),
]

pw_tmrca = []

for _ in tqdm(range(int(3e3))):
    ts = msprime.simulate(n, Ne=N, demographic_events=demographic_events)
    tree = ts.first()
    pw_tmrca.append(tree.tmrca(0, 1))

pw_tmrca = np.array(pw_tmrca)
```

Plot

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

x = np.sort(pw_tmrca)
y = np.arange(1, len(x) + 1) / len(x)
ax1.step(x, y, where="post",lw=3)
ax1.set_xlim(0,4*N)
ax1.axvline(x=T1, color='red', linestyle='--')
ax1.axvline(x=T2, color='red', linestyle='--')

times = [0, T1, T2, 4*N]
Pop_size = [N, N1, N2, N2]
ax2.step(times, Pop_size, where="post",lw=3)
ax2.set_xlim(0,4*N)
ax2.set_ylim(0,max(N,N1,N2)*1.1)
ax2.axvline(x=T1, color='red', linestyle='--')
ax2.axvline(x=T2, color='red', linestyle='--')

plt.show()
```
