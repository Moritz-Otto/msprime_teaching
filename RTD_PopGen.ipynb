```python
T1 = 1000
N1 = 500
T2 = 2000
N2 = 2000

demographic_events = [
    msprime.PopulationParametersChange(time=T1,initial_size=N1),
    msprime.PopulationParametersChange(time=T2,initial_size=N2),
]

iterations = 5000
tmrca = []
SNPs = []

for tm in range(iterations):
    ts = msprime.simulate(2,Ne=1e4,mutation_rate=1e-3,demographic_events=demographic_events)
    tree = ts.first()
    tmrca.append(tree.time(tree.root))
    SNPs.append((ts.genotype_matrix()).shape[0])
```

```python
plt.scatter(tmrca,SNPs,color="red",alpha=0.2,s=10)
plt.xlabel("TMRCA")
plt.ylabel("SNPs")
```

```python
%matplotlib inline
  
# getting data of the histogram
count, bins_count = np.histogram(tmrca, bins=100)
pdf = count / sum(count)
# finding the PDF of the histogram using count values
cdf = np.cumsum(pdf)

plt.plot(bins_count[1:], cdf, label="CDF")
plt.xlim(0,15000)
```
